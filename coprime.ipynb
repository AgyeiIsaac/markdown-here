```python
from math import gcd

# Function to generate all coprime pairs
def generate_coprime_pairs(start=2, end=1000):
    coprime_pairs = [
        (a, b)
        for a in range(start, end)
        for b in range(start, end)
        if gcd(a, b) == 1
    ]
    return coprime_pairs

# Generate the coprime pairs
pairs = generate_coprime_pairs()

# Print the number of pairs and some sample output
print(f"Total coprime pairs: {len(pairs)}")
print("All coprime pairs:", pairs[:1000])

# Optional: Save to a file
with open("coprime_pairs_2_to_1000.txt", "w") as f:
    for pair in pairs:
        f.write(f"{pair}\n")
```
